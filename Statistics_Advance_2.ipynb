**Question 1: What is hypothesis testing in statistics?**

**Answer**:- Hypothesis testing is a statistical method used to decide whether there is enough evidence in a sample to support a specific claim about a population.

● Null Hypothesis (H₀): The default assumption (e.g., no difference or effect).

● Alternative Hypothesis (H₁): What you want to test for (e.g., there is a difference or effect).

● You collect sample data, calculate a test statistic, and get a p-value.

● If the p-value is less than your chosen significance level (commonly 0.05), you reject H₀; otherwise, you fail to reject H₀. It's used to make informed decisions using sample data.

**Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Answer**:- The null hypothesis (H₀) is the default assumption that there is no effect, no difference, or no change in the population.

The alternative hypothesis (H₁ or Ha) is what you want to prove — it suggests there is an effect, a difference, or a change.

**Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

**Answer**:- In hypothesis testing, the significance level (denoted by α) is a critical threshold that helps you decide whether to reject the null hypothesis. It represents the probability of making a Type I error, which occurs when you reject a true null hypothesis.

The significance level (commonly 0.05, 0.01, or 0.10) is set before testing and defines the maximum acceptable risk of incorrectly rejecting a true null hypothesis.

**Step-by-step role (short format):**

1. **State Hypotheses**
* H₀: Default assumption (e.g., no effect)
* H₁: What you're testing for

2. **Set Significance Level (α)**
* e.g., α = 0.05 (5% risk of Type I error)

3. **Conduct Test**
* Compute test statistic and p-value

4. **Make Decision**
* If p ≤ α → Reject H₀
* If p > α → Fail to reject H₀

**Question 4: What are Type I and Type II errors? Give examples of each.**

**Answer**:-

1. **Type I Error (False Positive)**

* **Definition**: Rejecting the null hypothesis when it is actually true.

* **Symbol**: Probability = α (significance level)

* **Example**: A medical test shows a patient has a disease (rejects H₀: no disease), but the patient is actually healthy.

* **Real-life effect**: Unnecessary treatment or stress.

2. **Type II Error (False Negative)**

* **Definition**: Failing to reject the null hypothesis when it is actually false.
* **Symbol**: Probability = β
* **Example**: A medical test shows a patient is healthy (fails to reject H₀: no disease), but the patient actually has the disease.

* **Real-life effect**: Missed diagnosis, delayed treatment.

**Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.**

**Answer**:- A **Z-test** is used when you have a large sample size (n ≥ 30) and the population standard deviation (σ) is known. It relies on the normal distribution and is typically used for comparing sample and population means.

A **T-test** is used when you have a small sample size (n < 30) and the population standard deviation is unknown. It uses the t-distribution, which accounts for more variability in small samples, making it more appropriate for such cases.

In short:

● Use **Z-test** for large samples with known σ.

● Use **T-test** for small samples with unknown σ.

```python
#Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
import numpy as np
from scipy.stats import norm
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
mu_0 = 50
sigma = 0.5
alpha = 0.05
sample_mean = np.mean(sample_data)
n = len(sample_data)
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print(f"Sample mean: {sample_mean:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < alpha:
    print("Reject the null hypothesis (H₀).")
else:
    print("Fail to reject the null hypothesis (H₀).")
```

```python
#Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
np.random.seed(0)
mu = 100
sigma = 15
n = 100
data = np.random.normal(mu, sigma, n)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='blue', linestyle='dashed', linewidth=2, label=f'Mean = {sample_mean:.2f}')
plt.axvline(ci_lower, color='red', linestyle='dashed', linewidth=2, label=f'95% CI Lower = {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f'95% CI Upper = {ci_upper:.2f}')
plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt
def calculate_and_plot_z_scores(data):
  mean = np.mean(data)
  std = np.std(data, ddof=1)
  z_scores = (data - mean) / std
  plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
  plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
  plt.title('Histogram of Z-scores')
  plt.xlabel('Z-score (Standard Deviations from Mean)')
  plt.ylabel('Frequency')
  plt.legend()
  plt.grid(True)
  plt.show()
  return z_scores
sample_data = np.random.normal(loc=100, scale=15, size=100)
z_scores = calculate_and_plot_z_scores(sample_data)
```
